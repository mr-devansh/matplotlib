## OPTIONS OF SUBPLOTS

### OPTION 1

```python
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows = 2,
                                         ncols=2,
                                         figsize=(10,5))
ax1.plot(x,x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

### OPTION 2

```python
fig, ax = plt.subplots(nrows=2,
                      ncols=2,
                      figsize=(10,10))
ax[0,0].plot(x,x/2);
ax[0,1].scatter(np.random.random(10), np.random.random(10));
ax[1,0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1,1].hist(np.random.randn(1000));
```

### WHICH ONE TO USE (PYPLOT VS MATPLOTLIB OO)
* for quick plotting use pyplot
* for complex use matplotlib OO

```python
over_50 = heart_disease[heart_disease["age"]>50]
over_50
```

```python
# pyplot method
over_50.plot(kind="scatter",
            x="age",
            y="chol",
            c="target");
```

```python
# matplotlib OO method
fix, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter",
            x="age",
            y="chol",
            c="target",
            ax=ax);
ax.set_xlim([45,100])
```

```python
# purelymatllpotlib OO method 
fig, ax = plt.subplots(figsize=(10,6))
scatter = ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"])
ax.set(title="Heart Disease And Cholestrol",
      xlabel="Age",
      ylabel="Cholestrol")
ax.legend(*scatter.legend_elements(),title="Target");
ax.axhline(over_50["chol"].mean(),
          linestyle="--");
```

```python
# subplot of cholestrol and thalach with age
fig, (ax0,ax1) = plt.subplots(nrows=2,
                             ncols=1,
                             figsize=(10,10))
# ax0 defintions
scatter = ax0.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"])
ax0.set(title="Diseases",
       ylabel="Cholestrol")

ax.legend(*scatter.legend_elements(), title="Target")

ax0.axhline(y=over_50["chol"].mean(),
           linestyle="--");

scatter = ax1.scatter(x=over_50["age"],
                     y=over_50["thalach"],
                     c=over_50["target"])

# ax1 definitions
ax1.set(xlabel="Age",
       ylabel="Thalach")

ax1.legend(*scatter.legend_elements(), title="Target")

ax1.axhline(y=over_50["thalach"].mean(),
           linestyle="--");

# figure definitions
fig.suptitle("HEART DISEASE ANALYSIS", fontsize=16, fontweight="bold")
```
